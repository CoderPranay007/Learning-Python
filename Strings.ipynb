```python
print(Text.find('in'))

# Will tell the index of starting of this particular part of string

print(Text.find('is'))
```

```python
Text.replace('Australia', 'New Zealand')
print(Text)

# It will not work it is not returning value in place it is returning the new string
```

```python
New_text = Text.replace('Australia', 'New Zealand')
print(New_text)

#Here we set a new variable for a string
```

```python
#Instead we can set the same variable 

Text="I live in Australia, Australia is great"

print(Text)
print()
Text = Text.replace('Australia', 'New Zealand')
print(Text)
```
