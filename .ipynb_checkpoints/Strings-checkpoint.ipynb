```python
Text='Delhi is polluted'
print(len(Text))
# Length of a string
```

```python
Text='Delhi is polluted'

print(Text[10])
#Each character is stored at separate index
```

```python
Text="I live in Australia, Australia is great"

print(Text.lower())
# Converts entire text to lower case
```

```python
Text="I live in Australia, Australia is great"

print(Text.upper())
# Converts entire text to upper case
```

```python
print(Text.count('Australia'))
#Counts the exact same part of the string
```

```python
print(Text.find('in'))

# Will tell the index of starting of this particular part of string

print(Text.find('is'))
```

```python
Text.replace('Australia', 'New Zealand')
print(Text)

# It will not work it is not returning value in place it is returning the new string
```

```python
New_text = Text.replace('Australia', 'New Zealand')
print(New_text)

#Here we set a new variable for a string
```

```python
#Instead we can set the same variable 

Text="I live in Australia, Australia is great"

print(Text)
print()
Text = Text.replace('Australia', 'New Zealand')
print(Text)
```
